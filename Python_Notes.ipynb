### Estructura de datos

```python
# lista 
nums = [1,2,3,4]
nums.append(5)
nums[0]     # 1
nums[-1]    # 5

# lista como comprensión
squares = [x**2 for x in nums if x % 2 == 0]

# diccionario
person = {"name": "Ana", "age": 28}
person["age"]
person.get("age", "NA")

# lista como comprensión
squared = {x: x**2 for x in range(5)}

# tupla
point = (3, 4)
x, y = point

# sets
A = {1, 2, 3}
B = {3, 4, 5}
A & B   # intersección
A | B   # unión
```

### Control de flujo

```python
# Condicionales
if x > 0:
    print("positivo")
elif x == 0:
    print("cero")
else:
    print("negativo")
```

```python
# bucles
for i in range(5):
    print(i)

for i, val in enumerate(nums):
    print(i, val)
    
for a, b in zip([1,2], [3,4]):
    print(a + b)
```

### Funciones

```python
def mean(x):
    return sum(x) / len(x)

def power(x, p=2):
    return x**p
```
